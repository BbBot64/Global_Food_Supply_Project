```python
import folium
import pandas as pd
```

```python
my_map = folium.Map(location=[13.133932434766733, 16.103938729508073], zoom_start=1.4)
my_map
```

```python
df = pd.read_csv('Real_Final_Integrated_Data.csv')
country_df = df[['Country', 'Latitude', 'Longitude']]
country_df = country_df.drop_duplicates()
country_df
```

```python
country = country_df.loc[0]
folium.Marker(location=[country['Latitude'], country['Longitude']]).add_to(my_map)
my_map
```

```python
world_all_countries = folium.Map(location=[20, 10], zoom_start=2)
for row, country in country_df.iterrows():
    folium.Marker(
        location=[country['Latitude'], country['Longitude']], popup = country['Country'], tooltip = country['Country']).add_to(world_all_countries)
    
world_all_countries
```
