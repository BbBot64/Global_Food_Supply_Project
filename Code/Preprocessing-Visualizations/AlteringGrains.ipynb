```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('callback_grains.csv')
```

```python
df['Attribute_Description'].nunique()
```

```python
dfe = df.pivot(index = ['Commodity_Code', 'Commodity_Description', 'Country_Name', 'Market_Year'], columns = 'Attribute_Description', values = 'Value')
```

```python
dfe
```

```python
dfe['Exports'] *= 1000
dfe['Imports'] *= 1000
dfe['Area Harvested'] *= 1000
dfe['Beginning Stocks'] *= 1000
dfe['Ending Stocks'] *= 1000
dfe['FSI Consumption'] *= 1000
dfe['Feed Dom. Consumption'] *= 1000
dfe['Production'] *= 1000
dfe['TY Exports'] *= 1000
dfe['TY Imp. from U.S.'] *= 1000
dfe['TY Imports'] *= 1000
dfe['Milling Rate (.9999)'] *= 1000
dfe['Rough Production'] *= 1000
```

```python
dfe2 = dfe.drop(['Domestic Consumption', 'Total Distribution', 'Total Supply', 'Yield'], axis = 1)
```
